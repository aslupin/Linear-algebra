ในกิจกรรมที่สอง เราจะพิจารณาแนวคิดของ span  เราจะสนใจ vector ใน 2 และ 3 มิติ

```python
# ฟังก์ชันด่านล่างมาจากกิจกรรมที่แล้ว อย่าลืมคัดลอก vsub ที่เขียนมาด้วย

def scalar_mult(a, v):
    return [a*x for x in v]

def vadd(v1, v2):
    return [v1[i] + v2[i] for i in range(len(v1))]

def vsub(v1, v2):
    return [v1[i] - v2[i] for i in range(len(v1))]
```

```python
import matplotlib.pyplot as plt
def plot_vectors(vectors, scale):
    xcomp = [v[0] for v in vectors]
    ycomp = [v[1] for v in vectors]

    plt.quiver([0]*len(vectors), [0]*len(vectors), xcomp, ycomp, scale=scale)
    plt.show()
```

### >>> งาน 2.1 หาสัมประสิทธิ์
ให้ `v1` และ `v2` ดังต่อไปนี้

```python
v1 = [2,5]
v2 = [3,1]
```

ให้ `v3` ดังต่อไปนี้ ให้คำนวณหาค่าสัมประสิทธิ์ `c1` และ `c2` ที่ `c1 * v1 + c2 * v2` เท่ากับ `v3`

```python
v3 = [-10,7]
```

```python
# งานของคุณ ให้คำนวณหา c1 และ c2 จากนั้นนำมาป้อนที่นี้ และทดลอง plot เปรียบเทียบด้วยคำสั่งถัดไป
c1 = 31/13
c2 = 7-(155/13)
```

```python
vout = vadd(scalar_mult(c1,v1), scalar_mult(c2,v2))
plot_vectors([v3,vout], 30)
```

## Span ของเวกเตอร์ในสามมิติ
เราจะทดลองสร้าง linear combination ของเวกเตอร์ `v1` และ `v2` เพื่อพยายามจะมองภาพของ Span {v1, v2} ให้ออก  ให้ทดลองปรับเปลี่ยนค่า `v1` และ `v2` เพื่อดูผล

### >>> งาน 2.2 ฟังก์ชันคำนวณเซตของ linear combination

ให้นำโค้ดจากกิจกรรมที่แล้วที่คำนวณ linear combination ของเวกเตอร์สองตัว มาเขียนเป็นฟังก์ชันดังด้านล่าง เพื่อเตรียมไว้ใช้พล๊อตต่อไป  ฟังก์ชันนี้ควรทำงานได้กับเวกเตอร์ขนาดเท่าใดก็ได้

```python
def linear_comb(u1, u2):
    vectors = []
    for a in range(-10,10):
        for b in range(-10,10):
            # งานของคุณ: คำนวณ linear combination ของ v1 และ v2 โดยใช้ a/10 กับ b/10 เป็นสัมประสิทธิ์ ใส่ใน v
            # ใช้ scalar_mult กับ vadd ได้
            v = vadd(scalar_mult(a/10,v1),scalar_mult(b/10,v2))
            vectors.append(v)
    return vectors
```

```python
v1 = [1,2,1]
v2 = [0,1,1]

vects = linear_comb(v1,v2)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter([v[0] for v in vects], [v[1] for v in vects], [v[2] for v in vects])
```
